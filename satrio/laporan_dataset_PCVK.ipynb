# **Resolusi Citra**

```python
import os
from PIL import Image

folder_utama = '/content/drive/MyDrive/fruits-360/Training'

# Ekstensi file gambar yang akan dihitung
ekstensi_gambar = ('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff')

# Dictionary untuk menyimpan total resolusi
resolusi_data = {}

for nama_folder in sorted(os.listdir(folder_utama)):
    path_folder = os.path.join(folder_utama, nama_folder)
    if os.path.isdir(path_folder):
        total_width = 0
        total_height = 0
        jumlah_gambar = 0

        for nama_file in os.listdir(path_folder):
            if nama_file.lower().endswith(ekstensi_gambar):
                path_gambar = os.path.join(path_folder, nama_file)
                try:
                    with Image.open(path_gambar) as img:
                        width, height = img.size
                        total_width += width
                        total_height += height
                        jumlah_gambar += 1
                except Exception as e:
                    print(f"Gagal membaca {path_gambar}: {e}")

        if jumlah_gambar > 0:
            rata_rata_width = total_width / jumlah_gambar
            rata_rata_height = total_height / jumlah_gambar
            resolusi_data[nama_folder] = (jumlah_gambar, rata_rata_width, rata_rata_height)
            print(f"{nama_folder}: {jumlah_gambar} citra | Rata-rata resolusi: {rata_rata_width:.0f}x{rata_rata_height:.0f}")
```

# **Persebaran Warna Tiap Kelas**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os

# folder utama
parent_folder_path = "/content/drive/MyDrive/Colab Notebooks/PCVK/Collabs/fruit-clean/Test"

for folder_name in os.listdir(parent_folder_path):
    folder_path = os.path.join(parent_folder_path, folder_name)

    # Lewati jika bukan folder
    if not os.path.isdir(folder_path):
        continue

    print(f"Processing RGB folder: {folder_name}")

    all_r, all_g, all_b = [], [], []

    for filename in os.listdir(folder_path):
        if filename.lower().endswith((".jpg", ".jpeg", ".png")):
            img_path = os.path.join(folder_path, filename)

            img = cv2.imread(img_path)

            # Skip jika gambar gagal dibaca
            if img is None:
                print(f"  ❌ Gambar rusak atau tidak bisa dibaca: {filename}")
                continue

            # Jika RGBA → Konversi ke BGR biasa
            if img.shape[-1] == 4:
                img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)

            # Jika grayscale → jadikan RGB (duplikasi)
            if len(img.shape) == 2:
                img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

            rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Pecah channel
            r, g, b = cv2.split(rgb)

            all_r.extend(r.flatten())
            all_g.extend(g.flatten())
            all_b.extend(b.flatten())

    # Konversi ke array numpy
    all_r = np.array(all_r)
    all_g = np.array(all_g)
    all_b = np.array(all_b)

    # ==========================
    # Histogram RGB
    # ==========================
    plt.figure(figsize=(8,5))
    plt.title(f"Histogram Warna (RGB) - {folder_name}")
    plt.hist(all_r, bins=256, color="r", alpha=0.5, label="Red")
    plt.hist(all_g, bins=256, color="g", alpha=0.5, label="Green")
    plt.hist(all_b, bins=256, color="b", alpha=0.5, label="Blue")
    plt.xlabel("Intensitas")
    plt.ylabel("Jumlah Piksel")
    plt.legend()
    plt.xlim([0,256])
    plt.show()
```

# **Persebaran Cahaya**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python

# folder utama
parent_folder_path = "/content/drive/MyDrive/Colab Notebooks/PCVK/Collabs/fruit-clean/Test"

# Loop setiap folder kategori
for folder_name in os.listdir(parent_folder_path):
    folder_path = os.path.join(parent_folder_path, folder_name)

    if not os.path.isdir(folder_path):
        continue

    print(f"Processing Brightness (Luminance) folder: {folder_name}")

    all_light = []

    for filename in os.listdir(folder_path):
        if filename.lower().endswith((".jpg", ".jpeg", ".png")):
            img_path = os.path.join(folder_path, filename)

            # Baca gambar
            img = cv2.imread(img_path)
            rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Pecah channel
            R, G, B = cv2.split(rgb)

            # Hitung luminance
            L = 0.2126 * R + 0.7152 * G + 0.0722 * B

            # Simpan semua nilai kecerahan
            all_light.extend(L.flatten())

    all_light = np.array(all_light)

    # Histogram Luminance
    plt.figure(figsize=(8,5))
    plt.title(f"Histogram Pencahayaan (Luminance) - {folder_name}")
    plt.hist(all_light, bins=256, color="black", alpha=0.7)
    plt.xlabel("Intensitas Cahaya (0-255)")
    plt.ylabel("Jumlah Piksel")
    plt.xlim([0,256])
    plt.show()
```
